5.  	Which department is consuming most of the salary in comparison with other departments? Use seaborn’s catplot and use Gender as hue in this plot.


```python
sns.catplot(x="Department",y="Salary",data=hr,hue="Gender")
plt.gcf().set_size_inches(12, 8)
```

6.  	What is the average salary in each department?


```python
df = hr.groupby(['Department']).mean()
sns.barplot(x = df.index, y = 'Salary', data= df)
plt.gcf().set_size_inches(12,5)
```

7.  	Which employee has most absences in each department? Apply filter for absences > 19.


```python
df1 = hr[hr.Absences>19]
df1 = df1.groupby("Department").first() #select first of each department
sns.barplot(x="Employee_Name", y="Absences", data=df1)
plt.gcf().set_size_inches(12,5)
```

8.  	Which employee arrives late in last 30 days in each department? Apply filter for DaysLateLast30 > 3.

```python
df2 = hr[hr.DaysLateLast30>3]
df2 = df2.sort_values("Absences",ascending=False)
df2 = df2.groupby(["Department"]).first()
sns.barplot(x=df2.index,y="Absences",data=df2)
plt.gcf().set_size_inches(9,5)
```

9.  	Which employees in each department have performance score and engagement survey score of 5? Create bar plot and apply filter for Engagement Survey = 5 and Employee Satisfaction = 5.

```python
df3 = hr[(hr.EngagementSurvey==5) & (hr.EmpSatisfaction==5)]
sns.barplot(y="Employee_Name",x="EmpSatisfaction",data=df3,orient="h")
sns.barplot(y="Employee_Name",x="EngagementSurvey",data=df3,orient="h")
plt.gcf().set_size_inches(8,10)
```

10.  Which department is working best in terms of special project counts? i.e per employee projects count is > 6.


```python
df4 = hr[hr.SpecialProjectsCount>6]
df4 = df4.groupby(["Department"]).agg({"SpecialProjectsCount":"max"})
sns.barplot(x=df4.index,y="SpecialProjectsCount",data=df4)
plt.gcf().set_size_inches(3,4)
```

11.  Which employees have engagement score of 5 but also have maximum late days in last 30 days? Apply filter for Engagement Survey Score = 5.


```python
df5 = hr[hr.EngagementSurvey==5]
df5 = df5.loc[df5.DaysLateLast30==df5.DaysLateLast30.max()]
sns.barplot(x="Employee_Name",y="DaysLateLast30",data=df5)
plt.gcf().set_size_inches(3,4)
```

12.  Which employees are most deserving for promotion and incentives based on the following criteria:<br>
a.   	Special Project Counts > 4<br>
b.  	Employee Satisfaction Survey Score must be > 4.5


```python
df6 = hr[(hr.SpecialProjectsCount>4) & (hr.EmpSatisfaction>4.5)]
df6
sns.barplot(y=df6.Employee_Name,x=df6.EmpSatisfaction,orient="h")
```

13.  Which employees are worst employees based on the following criteria:
a.   	Performance Score: PIPv(Performance Improvement Plan)
b.  	Maximum Employee Satisfaction Survey Score must be < 3
c.   	Late in last 30 days: 3.

```python
df7 = hr[(hr.PerformanceScore=="PIP") & (hr.EmpSatisfaction<3.0)  & (hr.DaysLateLast30 >= 3)]
df7
sns.barplot(x=df7.EmpSatisfaction,y=df7.Employee_Name,orient="h")
```
